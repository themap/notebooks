```python
url="https://data.cdc.gov/data.json"
```

```python
response = urlopen(url)
  
data = json.loads(response.read())
  
#print(data)

for d in data['dataset']:
    print(d["title"])
    print(d["identifier"])
    print(d["landingPage"])
```
