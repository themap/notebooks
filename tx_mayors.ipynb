```python
from bs4 import BeautifulSoup
import urllib.request
import unidecode
import pandas as pd
import geopandas as gpd
```

```python
al = "https://directory.tml.org/results?search%5Btitle%5D%5B%5D=AAAA&search%5Bsubmit%5D=&search%5Btype%5D=title"
    
    
with urllib.request.urlopen(al) as response:
    html=response.read()

soup = BeautifulSoup(html)
soup



rows = soup.find_all("div", {"class": "tbody"})
for r in rows:
    name = r.findChildren("div", {"class": "td td-name"})
    mayor = name[0].findChildren("a")
    #print(mayor[0]["href"] )
    city = r.findChildren("div", {"class": "td td-city"})
    place = city[0].findChildren("a")
    print(place[0]["href"] + "|" + place[0].text.strip() + "|" + mayor[0]["href"] + "|" + mayor[0].text.strip())
```
