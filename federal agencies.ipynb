```python
from string import ascii_lowercase
letters=[]
for c in ascii_lowercase:
    letters.append(c)
letters
```

```python
all_links=[]

for l in letters:

    url = "https://www.usa.gov/federal-agencies/"+l + "#current-letter"

    with urllib.request.urlopen(url) as response:
        html=response.read()

    soup = BeautifulSoup(html)
    #print(soup)

    uls = soup.find_all("ul", {"class":"one_column_bullet"})



    # lis = soup.find_all("li")

    for u in uls:
        children = u.findChildren("li" , recursive=False)
        for c in children:
            childs = c.findChildren("a" , recursive=False)
            for d in childs:
                all_links.append("https://www.usa.gov" + d["href"])
            #print(d["href"])
    
    
    #     print(t)
#print(tds.findChildren)
    
    #print(t.findChildren()[0])

    
# tds = soup.find_all("a", {"class":"url"})
# for t in tds:
#     print(t)
```
