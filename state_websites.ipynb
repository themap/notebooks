```python
from bs4 import BeautifulSoup
import urllib.request
import unidecode
import pandas as pd
import geopandas as gpd
```

```python
url3="http://www.insidepolitics.org/states.html"
with urllib.request.urlopen(url3) as response:
    html2=response.read()

soup2 = BeautifulSoup(html2)

links = soup2.find_all("a")

for t in links:
    print(t.findChildren()[0].text + "|" + t["href"])
```
